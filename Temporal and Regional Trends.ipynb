```python
import pandas as pd
import numpy as np
import matplotlib as plot
```

```python
data = pd.read_csv("GenZ_DatingApp_Data.csv")
```

```python
# Are younger or older Gen-Z users more active on dating apps?

bins = [18, 22, 27]
labels = ["18-22 (Younger Gen-Z)", "23-27 (Older Gen-Z)"]
data["GenZ_Age_Group"] = pd.cut(data["Age"], bins=bins, labels=labels, right=True)
```

```python
age_group_usage_frequency = data.groupby(['Age', 'Usage_Frequency']).size().unstack(fill_value=0)
print(age_group_usage_frequency)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Plot the data
age_group_usage_frequency.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Dating App Usage Frequency by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Users')
plt.show()
```

```python
#Line Chart: Track trends over time
over_time = data.groupby(['Age', 'Primary_App'], observed=True).size().unstack().fillna(0)

plt.figure(figsize=(10, 6))
for app in over_time.columns:
    plt.plot(over_time.index, over_time[app], marker='o', label=app)
plt.title('Dating App Trends Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Users')
plt.legend(title='Primary App')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

city_app_usage = data.groupby(['Location', 'Primary_App']).size().unstack(fill_value=0)
sns.set(style="whitegrid")

# Plot the data
city_app_usage.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Dating App Usage Across Cities')
plt.xlabel('City')
plt.ylabel('Number of Users')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

app_popularity_by_age = data.groupby(['Age', 'Primary_App']).size().unstack(fill_value=0)

app_popularity_by_age.plot(kind='bar', figsize=(10, 6))
plt.title('App Popularity Across Age Groups')
plt.xlabel('Age')
plt.ylabel('Number of Users')
plt.legend(title='Primary App', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
